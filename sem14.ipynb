```python
s = [("Сарсик", 12, "Олег", "Ширяев"), ("Барсик", 8, "Алексей", "Пятницкий"), ("Пирожок", 4, "Олег", "Ширяев"), ("Мегатрон", 3, "Иван", "Захаров"), ("Оптимус прайм", 1, "Иван", "Захаров")]
dict = {}

for tuple in s:
    a,b,c,d = tuple 
    if c+" "+d in dict:
        dict[c+" "+d]+= a,b
    else:
        dict[c+" "+d] = a,b
print(dict)
```
